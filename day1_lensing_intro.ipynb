## Scientific Background

Gravitational lensing occurs when massive objects such as galaxy clusters bend the path of light from distant galaxies.

This effect arises from Einstein's theory of General Relativity, which states that mass curves spacetime. When light passes near a massive object, its trajectory bends along this curved spacetime.

By observing distortions in the shapes of background galaxies, astronomers can infer the mass distribution of the lensing object. This includes both visible matter and dark matter.

## Types of Gravitational Lensing

### Strong Lensing
- Produces dramatic distortions such as arcs and Einstein rings.
- Occurs when the alignment between the source, lens, and observer is very strong.

### Weak Lensing
- Causes small distortions in galaxy shapes.
- Requires statistical analysis of many galaxies.

### Microlensing
- Caused by smaller objects such as stars or planets.
- Often used in exoplanet detection.

## Research Notes

Important observables in gravitational lensing include:

### Shear (γ)
Describes the stretching of galaxy images due to gravitational lensing.

### Convergence (κ)
Represents the projected mass density along the line of sight.

### Magnification
Indicates the increase in brightness caused by lensing.

These quantities help reconstruct the mass distribution of galaxy clusters, including dark matter.
